Example 1 A Simple cost function
$f(x) = x^2 + x + 1$

```python
def f(x):
    return x**2 + x +1
```

```python
#make Data
x_1 = np.linspace(start = -3, stop = 3, num = 100)
print(x_1)
```

```python
#plot 
plt.xlim([-3, 3])
plt.ylim(0, 8)
plt.xlabel('X', fontsize = 16)
plt.ylabel('f(x)', fontsize = 16)
plt.plot(x_1, f(x_1))
plt.show()
```

## Slope and derivation
Challenge : create a python function for the derivative
of f(x) called df(x)

```python
def df(x):
    return 2*x + 1
```

## Subplot
plote function and derivative side by side



```python
#plot function and derivative side by side

plt.figure(figsize = [15,5])
#Chart 1: Cost function
plt.subplot(1,2, 1)
plt.title('slope of the cost function')
plt.xlim(-3, 3)
plt.ylim(0, 8)
plt.xlabel('X', fontsize = 16)
plt.ylabel('f(x)', fontsize = 16)
plt.plot(x_1, f(x_1),color = "blue", linewidth = '3')

#char 2: Chart derivative
plt.subplot(1,2,2)
plt.title('Slope of the cost function')
plt.ylabel('df(X)')
plt.xlabel('X')
plt.grid()
plt.xlim(-2, 3)
plt.ylim(-3, 6)
plt.plot(x_1, df(x_1), color = "skyblue", linewidth = '3')
```

# Gradient Descent

```python
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.00001

x_list = [new_x]
slope_list = [df(new_x)]



for n in range(500):
    previous_x = new_x
    gradient = df(previous_x)
    new_x = previous_x - step_multiplier * gradient
    
    step_size = abs(new_x - previous_x)
    #print(step_size)
    
    x_list.append(new_x)
    slope_list.append(df(new_x))
    
    if step_size < precision:
        print('the loop run this many times',n)
        break
    
print('Local minimum occures at :', new_x)
print('Slope or df(x) value at this point is: ', df(new_x))
print('f(x) value or cost at this point is:', f(new_x))
```

```python
#Superimpose the gradient descent calculations on plot 

plt.figure(figsize = [20,5])
#Chart 1: Cost function
plt.subplot(1,3, 1)
plt.title('Cost function', fontsize = 16)
plt.xlim(-3, 3)
plt.ylim(0, 8)
plt.xlabel('X', fontsize = 16)
plt.ylabel('f(x)', fontsize = 16)
plt.plot(x_1, f(x_1),color = "blue", linewidth = '3')

values = np.array(x_list)
plt.scatter(x_list, f(values),color = 'red', s = 100, alpha = 0.6)

#char 2: Chart derivative
plt.subplot(1,3,2)
plt.title('Slope of the cost function', fontsize = 16)
plt.ylabel('df(X)')
plt.xlabel('X')
plt.grid()
plt.xlim(-2, 3)
plt.ylim(-3, 6)
plt.plot(x_1, df(x_1), color = "skyblue", linewidth = '3',alpha = 0.6)
plt.scatter(x_list, slope_list, color = 'red', s=100, alpha =0.5)

#Chat 3: Derivative (close up)
plt.subplot(1,3,3)
plt.title('Gradient Descent (close up)', fontsize=16)
plt.xlabel('X', fontsize = 16)
plt.grid()
plt.xlim(-0.55, -0.2)
plt.ylim(-0.3, 0.8)
plt.plot(x_1, df(x_1), color = 'skyblue', linewidth = 6, alpha = 0.8)
plt.scatter(x_list, slope_list, color = 'red', s =300, alpha = 0.6)

plt.show()
```

# Example 2- Multiple Minima vs Initial Guess & Advanced Functions

## $$g(x) = x^4 - 4x^2 + 5$$

```python
#Make some data
x_2 = np.linspace(-2, 2 , 1000)
#Challenge: write the g(x) function and the dg(x) function in python?

#g(x) function
def g(x):
    return x**4 - 4*x**2 + 5


#dg(x) function
def dg(x):
    return 4*x**3 - 8*x
```

```python
#plot function and derivative side by side

plt.figure(figsize = [15,5])
#Chart 1: Cost function
plt.subplot(1,2, 1)
plt.title('Cost function', fontsize = 16)
plt.xlim(-2, 2)
plt.ylim(0.5, 5.5)
plt.xlabel('X', fontsize = 16)
plt.ylabel('g(x)', fontsize = 16)
plt.plot(x_2, g(x_2),color = "blue", linewidth = '3')

#char 2: Chart derivative
plt.subplot(1,2,2)
plt.title('Slope of the cost function', fontsize = '16')
plt.ylabel('dg(X)', fontsize = 16)
plt.xlabel('X', fontsize = 16)
plt.grid()
plt.xlim(-2, 2)
plt.ylim(-6, 8)
plt.plot(x_2, dg(x_2), color = "skyblue", linewidth = '3')
plt.show()
```

# Gradient Descent as a Python Function

```python
def gradient_descent(derivative_func, initial_guess, multiplier = 0.02, precision = 0.001):
    #ctrl + ] for indentation of a multiples lines of code
    
    
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]



    for n in range(500):
        previous_x = new_x
        gradient =derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)
        #print(step_size)

        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        if step_size < precision:
            break
    return new_x, x_list, slope_list
```

```python
local_min, list_x, deriv_list = gradient_descent(dg, 0.5, 0.02, 0.001)
print('Local min occurs at:', local_min)
print('Number of steps:', len(list_x))
```

```python
local_min, list_x, deriv_list = gradient_descent(derivative_func = dg, initial_guess = -0.5, multiplier = 0.01,
                                                 precision = 0.0001)
print('Local min occurs at:', local_min)
print('Number of steps:', len(list_x))
```

```python
local_min, list_x, deriv_list = gradient_descent(derivative_func = dg, initial_guess = -0.1)
print('Local min occurs at:', local_min)
print('Number of steps:', len(list_x))
```

```python
#calling gradient descent function
local_min, list_x, deriv_list = gradient_descent(derivative_func = dg, initial_guess = 0.1)
#plot function and derivative and scatter plot side by side

plt.figure(figsize = [15,5])
#Chart 1: Cost function
plt.subplot(1,2, 1)
plt.title('Cost function', fontsize = 16)
plt.xlim(-2, 2)
plt.ylim(0.5, 5.5)
plt.xlabel('X', fontsize = 16)
plt.ylabel('g(x)', fontsize = 16)
plt.plot(x_2, g(x_2),color = "blue", linewidth = '3', alpha = 0.8)
plt.scatter(list_x, g(np.array(list_x)), color = 'red', s = 100, alpha = 0.6)

#char 2: Chart derivative
plt.subplot(1,2,2)
plt.title('Slope of the cost function', fontsize = '16')
plt.ylabel('dg(X)', fontsize = 16)
plt.xlabel('X', fontsize = 16)
plt.grid()
plt.xlim(-2, 2)
plt.ylim(-6, 8)
plt.plot(x_2, dg(x_2), color = "skyblue", linewidth = '5', alpha = 0.6)
plt.scatter(list_x, deriv_list, color = 'red', s = 100, alpha = 0.5)
plt.show()
```
